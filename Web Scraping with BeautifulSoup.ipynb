```python

import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
URL = "https://en.wikipedia.org/wiki/List_of_largest_banks"
html_data = requests.get(URL)
```

```python
html_data[101:124]
```

```python
soup = BeautifulSoup(html_data.content, 'html.parser')
```

```python
soup.title.text
```

```python
soup.prettify()
```

```python

Empty_list = []
for row in soup.find_all('tbody')[3].find_all('tr'):
    col = row.find_all('td')
    col=[x.text.strip() for x in col]
    print(col)
    Empty_list.append(col)
    
```

```python
data = pd.DataFrame(Empty_list, columns=["Rank", "Name", "Market Cap (US$ Billion)"])
data
```
